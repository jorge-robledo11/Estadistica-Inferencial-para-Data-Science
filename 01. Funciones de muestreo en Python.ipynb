### 2. Muestreo sistemático

```python
from pandas import DataFrame

def systematic_sampling(data:DataFrame, step:int):

    indexes = np.arange(0, len(data), step=step)
    systematic_sample = data.iloc[indexes]
    return systematic_sample


systematic_sample = systematic_sampling(data=data, step=3)
systematic_sample
```

### 3. Muestreo estratificado

```python
data['estratificado'] = data['delegacion'] + ',' + data['tipo']
(data['estratificado'].value_counts()/len(data)).sort_values(ascending=False)
```

```python
"""
Queremos estratificar en las siguientes proporconiones
1. Hoteles en Cauhtemoc: 0.5
2. Museos en Cauhtemoc: 0.2
3. Hoteles en Venustiana Carranza: 0.1
4. Mercados en Cauhtemoc: 0.1
5. Mercados en Venustiana Carranza: 0.1
"""

# Función para estratificar los valores
def data_stratified(data:DataFrame, names:str, values_stratify:list, prop_stratify:list, random_state=None):

    data_stratify = pd.DataFrame(columns=data.columns)

    pos = -1
    for i in range(len(values_stratify)):
        pos += 1
        if pos == len(values_stratify) -1:
            ratio_len = len(data) - len(data_stratify)
        else:
            ratio_len = int(len(data) * prop_stratify[i])

        data_filtered = data[data[names] == values_stratify[i]]
        data_temp = data_filtered.sample(replace=True, n=ratio_len, random_state=random_state)
        data_stratify = pd.concat([data_stratify, data_temp])
    
    return data_stratify


valores_estratificar = ['Cuautémoc,Hotel', 
                        'Cuautémoc,Museo', 
                        'Venustiano Carranza,Hotel',
                        'Cuauhtémoc,Mercado',
                        'Venustiano Carranza,Mercado']

proporcion_deseada = [0.5, 0.2, 0.1, 0.1, 0.1]

data_balanced = data_stratified(data=data, names='estratificado', values_stratify=valores_estratificar, prop_stratify=proporcion_deseada, random_state=1337)
data_balanced.head()
```

```python
# Proporciones de datos estratificados
(data_balanced.groupby('estratificado').size().sort_values(ascending=False))/len(data_balanced)
```
